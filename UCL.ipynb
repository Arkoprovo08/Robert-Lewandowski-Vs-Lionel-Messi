```python
cols_for_radar = ['accuratePassesPercentage','keyPasses','successfulDribblesPercentage','accurateCrossesPercentage','goalConversionPercentage','accurateLongBallsPercentage','minutesPlayed']
```

```python
df_1[cols_for_radar]
```

```python
df_final = pd.concat([df_1,df_2])
```

```python
df_final
```

```python
df_final[cols_for_radar]
```

```python
df_final['shotsper90'] = df_final['totalShots'] / df_final['minutesPlayed']*90
```

```python
df_final['shotsper90']
```

```python
df_final['bigChancesCreated'] = df_final["bigChancesCreated"]*10
```

```python
cols_for_radar = ['bigChancesCreated','accuratePassesPercentage','keyPasses','successfulDribblesPercentage','accurateCrossesPercentage','goalConversionPercentage','accurateLongBallsPercentage']
```

```python
radar_df = df_final.groupby(["name"])[cols_for_radar].sum().reset_index()
```

```python
# Initiate the plotly go figure
fig = go.Figure()
# Add Radar plots for different players:
fig.add_trace(go.Scatterpolar(
    r=radar_df.loc[(radar_df["name"] == "Robert Lewandowski"), cols_for_radar].sum(),
    theta=cols_for_radar,
    fill='toself',
    name='Robert Lewandowski'))
fig.add_trace(go.Scatterpolar(
    r=radar_df.loc[(radar_df["name"] == "Lionel Messi"), cols_for_radar].sum(),
    theta=cols_for_radar,
    fill='toself',
    name="Lionel Messi"))
# Additional properties for the plot:
fig.update_layout(
    title="Robert Lewandowski vs Messi 2020-21 UCL",
polar=dict(
    radialaxis=dict(
    visible=True,
    )),
showlegend=True
)
fig.show()
```
